```python
import sys
import os
sys.path.append(os.path.abspath(os.path.join('..', 'src')))
```

```python
from data_preprocessing import preprocess_data
from model_building import build_random_forest,build_linear_regression,build_decision_tree
from model_evaluations import evaluate_model
```

```python
filepath = '../data/raw/house_prices.csv'
target_column = 'SalePrice'
```

```python
X_train, X_test, y_train, y_test = preprocess_data(filepath, target_column)
```

```python
# Train the models
rf_model = build_random_forest(X_train, y_train)
dt_model = build_decision_tree(X_train, y_train)
lr_model = build_linear_regression(X_train, y_train)
```

```python
# Evaluate the models
rf_mse, rf_r2 = evaluate_model(rf_model, X_test, y_test)
dt_mse, dt_r2 = evaluate_model(dt_model, X_test, y_test)
lr_mse, lr_r2 = evaluate_model(lr_model, X_test, y_test)
```

```python
# Print evaluation results
print(f"Random Forest MSE: {rf_mse}, R²: {rf_r2}")
print(f"Decision Tree MSE: {dt_mse}, R²: {dt_r2}")
print(f"Linear Regression MSE: {lr_mse}, R²: {lr_r2}")
```
